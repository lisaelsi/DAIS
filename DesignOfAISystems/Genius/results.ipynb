```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, f1_score, accuracy_score
import matplotlib.pyplot as plt

# 0 - eminem
# 1 - ai


part1_correct = [0, 1, 1, 1, 0, 0]
part2_correct = [0, 1, 1, 0]

#correct = part1_correct + part2_correct

part1_predictions = [0, 1, 1, 1, 0, 0]

cm = confusion_matrix(part1_correct, part1_predictions)
disp = ConfusionMatrixDisplay(cm, display_labels = ['Eminem', 'AI'])
disp.plot()
plt.show()
```

```python
import pandas as pd

ans = pd.read_csv('./answers.csv')
ans.pop('Tidst채mpel')

replacements = {'AI': 1, 'Eminem': 0}
ans = ans.replace(to_replace=replacements)
```

```python
matrix = ans.values
answers = matrix.reshape(-1)
all_answers = list(answers)

all_correct = part1_correct + part2_correct
all_correct = all_correct * 23

cm = confusion_matrix(all_correct, all_answers)
disp = ConfusionMatrixDisplay(cm, display_labels = ['Eminem', 'AI'])
disp.plot()
plt.show()

f1 = f1_score(all_correct, all_answers)
print(f'F1 score for total (single and multiple lines): {f1}')

acc = accuracy_score(all_correct, all_answers)
print(f'Accuracy score for part one (single lines): {acc}')
```
